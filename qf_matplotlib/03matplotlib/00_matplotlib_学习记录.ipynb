```python
import numpy as np
import matplotlib.pyplot as plt
```

# 一 matplotlib 基础

```python
# 可以使用多个plot函数, 在一个图中绘制多个曲线(推荐)
x = np.linspace(1, 10, 10)
plt.plot(x)
plt.plot(x, x ** 2)
plt.show()
```

```python
# 也可以在一个plot函数中传入多对x, y值, 在一个图中绘制多个曲线
x = np.arange(0, 10, 1)
plt.plot(x, x ** 2, x, 5 * x, x, x / 3)
plt.grid(True)  # 添加网格线
plt.show()
```

```python
# 绘制正弦曲线
# lw代表linewidth, 线的粗细
# alpha表示线的明暗程度(0-1)
# color表示线的颜色
x = np.arange(-np.pi, np.pi, 0.01)
plt.plot(x, np.sin(x), x, np.cos(x), lw=10, alpha=0.5, color="blue")
plt.grid(True)
plt.show()
```

# 二 面向对象与设置标题

### plt面向对象, 图形就是所谓的对象

```python
# 设置整张图片的大小
plt.figure(figsize=(5, 3))

x1 = np.linspace(-np.pi, np.pi, 100)
plt.plot(x1, np.sin(x1))
# loc值: 'center', 'right', 'left'
plt.title("f(x) = sin(x)", fontsize=20, loc="right")
plt.show()
```

```python
# 设置整张图片的大小
plt.figure(figsize=(10, 7))

x1 = np.arange(-10, 10, 0.1)
# 图片显示的位置subplot(m行, n列, local最大值为m*n
axes1 = plt.subplot(2, 3, 1)
# 设置网格线的颜色, 线条的格式, 线条的宽度
axes1.grid(color="r", linestyle="-.", linewidth=2)
# 设置图片1标题字体大小与格式
axes1.set_title('sin', fontsize='large', fontweight='bold') 
axes1.plot(x1, np.sin(x1))

axes2 = plt.subplot(2, 3, 2)
x2 = np.arange(-10, 10, 0.1)
axes2.grid(color="g", linestyle="--", linewidth=1)
axes2.set_title('cos', fontsize='large', fontweight='bold')  
axes2.plot(x2, np.cos(x2))

axes3 = plt.subplot(2, 3, 3)
x3 = np.arange(-10, 10, 0.1)
axes3.grid(color="b", linestyle="-.", linewidth=1)
axes3.set_title('y = x ** 2', fontsize='large', fontweight='bold')  
axes3.plot(x3, x3 ** 2)

axes4 = plt.subplot(2, 3, 5)
x4 = np.arange(-10, 10, 0.1)
axes4.grid(color="gray", linestyle="-.", linewidth=1)
axes4.set_title('y = 4x + 1', fontsize='large', fontweight='bold')  
axes4.plot(x4, 4 * x4 + 1)

# 设置整张图片的图片名字
plt.suptitle('Picture', fontsize='large', fontweight='bold')

plt.show()
```

# 三 坐标轴界限

### 3-1 axis 方法:  tight(默认), off, equal, square

```python
# 设置整张图片的大小
plt.figure(figsize=(5, 3))
ax = plt.subplot(1, 2, 1)

# axis square/将当前坐标系图形设置为方形。横轴及纵轴比例是1：1
# axis equal/将横轴纵轴的定标系数设成相同值。
# 如果axis方法没有任何参数, 则返回当前坐标轴的上下限, axis(xmin=, ymax=)
x1 = np.random.rand(10)
ax.plot(x1)
ax.axis([-5, 15, -5, 10])  # [xmin, xmax, ymin, ymax]
ax.axis("off")  # 关闭坐标轴

ax2 = plt.subplot(1, 2, 2)
x2 = np.linspace(-np.pi, np.pi, 100)
ax2.plot(np.sin(x2), np.cos(x2))
ax2.axis("equal")  # 将横轴纵轴的定标系数设成相同值
ax2.axis("off")  # 关闭坐标轴
plt.show()
```

### 3-2 xlim方法与ylim方法

```python
plt.figure(figsize=(5, 3))
# xlim方法与ylim方法
y = np.arange(0, 10, 1)
plt.plot(y)

plt.ylim(2, 10)  # 从2开始到10结束
plt.xlim(4, 8)  # 从4开始到8结束
plt.show()
```

### 3-3 坐标轴标签  xlabel方法和ylable方法

```python
ax4 = plt.subplot(1, 2, 2)
x = np.arange(-5, 5, 0.1)
y = x ** 2 + 5
ax4.plot(x, y)

# horizontalalignment值: 'center', 'right', 'left'
plt.ylabel("f(x) = x**2 + 5", horizontalalignment="left")
# verticalalignment值: 'top', 'bottom', 'center', 'baseline', 'center_baseline'
# rotation表示倾斜角度
plt.xlabel("this is x", size=10, rotation=0, verticalalignment="top")
plt.show()
```

# 四 图例

### 4-1 legend

```python
"""图例(legend方法)
两种传参方法:
1. 在plot函数中增加label参数(推荐)
2. 在legend方法中传入字符串列表
"""
plt.figure(figsize=(10, 4))  # 设置整张图片的大小

# 方法一
ax1 = plt.subplot(1, 3, 1)
ax1.set_title("first")
x1 = np.arange(0, 10, 1)
ax1.plot(x1, x1, label="normal")
ax1.plot(x1, x1 ** 2, label="_fast")  # label添加下划线_, 被隐藏
ax1.plot(x1, x1 / 2, label="slow")
"""loc 位置值 
'best': 0,         'upper right': 1,    'upper left': 2,
'lower left': 3,   'lower right': 4,    'right': 5, 
'center left': 6,  'center right': 7,   'lower center': 8,
'upper center': 9, 'center': 10,
"""
ax1.legend()
# ax3.legend(loc="right")
# ax3.legend(loc=(-0.2, 1))
# ax3.legend(loc=(0.5, 1))
# ax1.legend(loc=(1, 0))

# 方法二
ax2 = plt.subplot(1, 3, 2)
ax2.set_title("second")
x2 = np.arange(0, 10, 1)
ax2.plot(x2, x2, x2, x2 * 2, x2, x2 / 2)
ax2.legend(["normal", "fast", "slow"])

# ncol 参数: 每一行显示几个
ax3 = plt.subplot(1, 3, 3)
ax3.set_title("third")
x3 = np.arange(0, 10, 1)
ax3.plot(x3, x3, x3, x3 * 2, x3, x3 / 2)
ax3.legend(["normal", "fast", "slow"], loc = 0, ncol = 2)
plt.show()
```

### 4-2 线条的颜色color, 格式linestyle, 标记marker

```python
plt.figure(figsize=(8, 4))
x1 = np.random.randn(20)
x2 = np.random.randn(20)
x3 = np.random.randn(20)
plt.plot(x1.cumsum(), color = "red", linestyle = "--", marker = "o")
plt.plot(x2.cumsum(), color = "yellow", linestyle = "-.", marker = ">")
plt.plot(x3.cumsum(), color = "purple", linestyle = ":", marker = "|")
plt.legend(["x1", "x2", "x3"])
plt.show()
```
